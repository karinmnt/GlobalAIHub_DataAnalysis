# Importing random nan values into data

This code defines a function to introduce random missing values (NaN) in a DataFrame. The function is customizable in terms of the percentage of missing data it should introduce.

```python
def add_random_missing_values(dataframe: pd.DataFrame,
                              missing_rate: float = 0.05) -> pd.DataFrame:
    """Turns random values to NaN in a DataFrame.

    To use this function, you need to import pandas, numpy and random libraries.

    Args:
        dataframe (pd.DataFrame): DataFrame to be processed.
        missing_rate (float): Percentage of missing value rate in float format. Defaults 0.05

    Returns:
        df_missing (pd.DataFrame): Processed DataFrame object.

    """
    # Get copy of dataframe
    df_missing = dataframe.copy()

    # Obtain size of dataframe and number total number of missing values
    df_size = dataframe.size
    num_missing = int(df_size * missing_rate)

    # Get random row and column indexes to turn them NaN
    for _ in range(num_missing):
        row_idx = random.randint(0, dataframe.shape[0] - 1)
        col_idx = random.randint(0, dataframe.shape[1] - 1)

        df_missing.iat[row_idx, col_idx] = np.nan

    return df_missing


df = add_random_missing_values(dataframe = df_copy,
                               missing_rate = 0.03)
```

```python
# Replacing missing CustomerIDs with random numbers

missing_ids = df['CustomerID'].isna()
df.loc[missing_ids, 'CustomerID'] = range(10000, 10000 + missing_ids.sum())
```

**missing_ids = df['CustomerID'].isna()**: This line creates a Boolean Series missing_ids that is True where the values in the CustomerID column are missing (NaN), and False where they are not.

**df.loc[missing_ids, 'CustomerID']:** This accesses the rows where missing_ids is True and selects the CustomerID column.

**range(10000, 10000 + missing_ids.sum()):** This generates a sequence of numbers starting from 10000, with a length equal to the number of missing values in the CustomerID column (missing_ids.sum() gives the total number of True values in missing_ids).

*The assignment df.loc[missing_ids, 'CustomerID'] = range(10000, 10000 + missing_ids.sum()) replaces the missing CustomerID values with this generated sequence.*

Since CustomerID should be specifc to customers instead of replacing it wih exisiting values, it is prefered to generate random IDs.

```python
# Replacing missing Transaction_IDs with random numbers

missing_ids_TS = df["Transaction_ID"].isna()
df.loc[missing_ids_TS, "Transaction_ID"] = range(10000, 10000 + missing_ids_TS.sum())
```

# Feature Engineering

Feature engineering is the process of selecting, modifying, or creating new features (variables) from raw data to improve the performance of a machine learning model. It is a critical step in the data preprocessing phase of a machine learning pipeline. The goal of feature engineering is to enhance the predictive power of the model by making the data more suitable for the machine learning algorithm.

```python
# Summarize total spend by category
category_spend_summary = df_cleaned.groupby('Product_Category')['Category_Spend'].first().reset_index()

print(category_spend_summary.head())
```

This shows how much is spent on each product category and can include the impact of different product categories in the model.

```python
# Adding the Coupon_Status column: 1 if Coupon_Code exists, 0 otherwise
df_cleaned['Coupon_Status'] = df_cleaned['Coupon_Code'].apply(lambda x: 1 if pd.notnull(x) else 0)

# Adding the Discount_Amount column: Calculating the discount amount
df_cleaned['Discount_Amount'] = df_cleaned['Total_Spend'] * (df_cleaned['Discount_pct'] / 100)

df_cleaned[['CustomerID', 'Coupon_Code', 'Discount_pct', 'Total_Spend', 'Coupon_Status', 'Discount_Amount']].head()
```

These features help to determine when users are using coupons and making discounted purchases.

```python
total_quantity = df_cleaned.groupby('CustomerID')['Quantity'].sum()
df_cleaned['Total_Quantity'] = df_cleaned['CustomerID'].map(total_quantity)

df_cleaned[['CustomerID', 'Total_Quantity', 'Quantity']].head()
```

This feature helps to understand the purchasing frequency of users.

```python
df_cleaned['Total_Offline_Online_Spend'] = df_cleaned['Offline_Spend'] + df_cleaned['Online_Spend']

df_cleaned[['Total_Offline_Online_Spend',"CustomerID"]].tail()
```

This can help to analyze the total amount spent by combining users' online and offline purchases.

```python
category_avg_price = df_cleaned.groupby('Product_Category')['Avg_Price'].mean()
df_cleaned['Category_Avg_Price'] = df_cleaned['Product_Category'].map(category_avg_price)

df_cleaned[["Product_Category",'Category_Avg_Price',"CustomerID"]].tail()
```

This process calculates the average price for each product category and allows to analyze user interaction with prices across product categories.
